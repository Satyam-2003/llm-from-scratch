```python
import torch
import torch.nn as nn
```

```python
vocab_size = 10
output_dim = 4

embedding = nn.Embedding(vocab_size, output_dim)
```

```python
input_ids = torch.tensor([2,5,7])
```

```python
embeddings = embedding(input_ids)
print(embeddings)
```

```python
input_ids.shape
embeddings.shape
```

```python
# call the embedding layer (callable) instead of the tensor 'embeddings'
print(embedding(torch.tensor([3])))
```
